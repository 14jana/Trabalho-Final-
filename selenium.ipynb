```python
import pandas as pd
import sqlalchemy as sqa
from selenium import webdriver
from selenium.webdriver.common.by import By

```

```python
navegador = webdriver.Chrome()
```

```python
navegador.get('https://pt.wikipedia.org/wiki/Lista_de_filmes_de_maior_bilheteria')
```

```python
navegador.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/div[2]').click()
```

```python
bilheteria = navegador.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/thead/tr/th[6]').text
ranking = navegador.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr[1]/td[1]').text
ano = navegador.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr[1]/td[2]').text 
filme = navegador.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr[1]/td[3]/i/a').text
diretor = navegador.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr[1]/td[5]').text
distribuidor = navegador.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[1]/tbody/tr[1]/td[4]/a').text


```

```python
print(bilheteria)
print(ranking)
print(ano)
print(filme)
print(diretor)
print(distribuidor)

```
